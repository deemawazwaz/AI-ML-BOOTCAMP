## Step 2: Handling Missing Values

We check for missing values in each column and clean the dataset. Missing values in Age are filled with the mean because age is numeric. Since Cabin contains too many missing values, we remove the column.

```python
df.isnull().sum()
```

```python
df["Fare"] = df["Fare"].fillna(df["Fare"].mean())
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df.drop(columns=["Cabin"], inplace=True)
```

```python
df.isnull().sum()
```

## Step 3: Filtering Data

We filter the dataset to show passengers older than 30 years old and examine the result.

```python
older_than_30 = df[df["Age"] > 30]
older_than_30.head()
```

```python
print("Number of passengers older than 30:", len(older_than_30))
```

The filtered dataset contains only passengers older than 30 years. This allows us to analyze a specific age group separately.

## Step 4: Grouping and Aggregation

We calculate the average fare for each passenger class using the groupby() function.

```python
df.groupby("Pclass")["Fare"].mean()
```

```python
df.groupby("Sex")["Age"].mean()
```

```python
df.groupby("Embarked").agg({
    "Age": "mean",
    "Fare": "mean"
})
```

### Interpretation

The groupby() operation shows how statistics differ between categories. For example, passenger classes have different average fares, and the average age can vary between groups.
