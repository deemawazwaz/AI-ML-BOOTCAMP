## Functions

Functions allow us to write reusable code. Instead of repeating the same logic multiple times, we define it once and call it whenever needed.

```python
def Spain(Loser):
    print("The Winner Takes It All!", Loser)

Spain("Congratulation SPAIN!!")   
Spain("1 - 0")   

```

```python
def Normalize(Value, Min_Value, Max_Value):
    return (Value - Min_Value)/(Max_Value - Min_Value)
print (Normalize(90,100,60))   
    
```

```python
def avg(f_mark, se_mark, th_mark, ft_mark, num_of_marks):
    return (f_mark + se_mark + th_mark + ft_mark)/ (num_of_marks)
print (avg(90,96,95,97,4))    
```

## Array Attributes

NumPy arrays have useful attributes such as shape, data type (dtype), and number of dimensions (ndim).

## Vectorized Operations

NumPy performs mathematical operations on entire arrays without using loops. This makes the code faster and easier to read.

```python
a = np.array([1, 2, 3, 4])

print("Original:", a)
print("Multiply by 2:", a * 2)
print("Add itself:", a + a)
print("Mean:", a.mean())
```

## Boolean Masking

Boolean masking filters array elements based on a condition.

```python
a = np.array([1, 9, 3, 4, 0 ])

print(a > 2)
print(a[a>2])
```

## Broadcasting

Broadcasting allows NumPy to perform operations between arrays of compatible shapes.

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

row = np.array([10, 20, 30])

print(matrix + row)
```

# Hands-On Lab

In this lab, we practice creating NumPy arrays, slicing, boolean masking, and broadcasting.

```python
array = np.arange(1, 17).reshape(4, 4)

print(array)
print("Shape:", array.shape)
print("Data Type:", array.dtype)
```

## Step 2: Array Slicing

Extract the second column and the last row using slicing.

```python
print("Second Column:")
print(array[:, 1])

print("Last Row:")
print(array[3, :])
```

## Step 3: Boolean Masking

Select all values greater than the mean of the array.

```python
mean = array.mean()

print("Mean:", mean)

print(array[array > mean])
```
