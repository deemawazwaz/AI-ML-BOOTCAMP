## Control Flow

Control flow allows Python to make decisions and repeat tasks using conditions and loops.

```python
score = 98
if score >= 60:
    print ("Pass, What A LEGEND!!")
else:
    print ("fail")
```

```python
Num = 1

while Num <= 10:
    print(Num)

    if Num % 2 == 0:
        print("Number is Even")
    else:
        print("Number is Odd")

    Num += 1
```

## Functions

Functions allow us to write reusable code. Instead of repeating the same logic multiple times, we define it once and call it whenever needed.

```python
def Spain(Loser):
    print("The Winner Takes It All!", Loser)

Spain("Congratulation SPAIN!!")   
Spain("1 - 0")   

```

```python
def Normalize(Value, Min_Value, Max_Value):
    return (Value - Min_Value)/(Max_Value - Min_Value)
print (Normalize(90,100,60))   
    
```

```python
def avg(f_mark, se_mark, th_mark, ft_mark, num_of_marks):
    return (f_mark + se_mark + th_mark + ft_mark)/ (num_of_marks)
print (avg(90,96,95,97,4))    
```
